<h1>Pre-procesamiento de los datos</h1>

```python
import pandas as pd
import datetime as dt
df = pd.read_csv('combined_dataSIN5.csv')
df_reducido = df.sample(frac=0.1, random_state=1) 
df_reducido.head(5)
```

<h1>Configuracion de Dates para calculo de la duración</h1>

```python
df_reducido['Inicio'] = pd.to_datetime(df_reducido['Fecha_Inicio'] + ' ' + df_reducido['Hora_Inicio'])
df_reducido['Final'] = pd.to_datetime(df_reducido['Fecha_Final'] + ' ' + df_reducido['Hora_Final'])
df_reducido = df_reducido.drop(columns=['Fecha_Inicio', 'Hora_Inicio','Fecha_Final','Hora_Final'])
df_reducido.head(5)
```

<h1>Primera dimension a añadir: Duracion</h1>

```python
#Los camentarios son hechos por mi, no por chatgpt
# Suponiendo que df es tu DataFrame y ya has creado las columnas 'Inicio' y 'Final'
df_reducido.loc[:, 'Duracion_Minutos'] = (df_reducido['Final'] - df_reducido['Inicio']).dt.total_seconds() / 60
df_reducido = df_reducido[(df_reducido['Duracion_Minutos'] >= 5) & (df_reducido['Duracion_Minutos'] <= 180)]# Ajustamos para que solo aparezcan procesos en ese intervalo
df_reducido
```
